```python

import pandas as pd
import plotly.plotly as py
import plotly.offline as po
import plotly.graph_objs as pg
```

```python
po.init_notebook_mode(connected=True)
```

```python
corona= pd.read_csv('corona_country_wise_latest.csv')
corona
```

```python
data = dict(type = 'choropleth',
            locations = corona['Country/Region'],
            locationmode = 'country names',
           colorscale = 'Portland',
            text= "Active Corona Case",
            z=corona['Active'],
            colorbar = {'title':'Corona Active Cases', 'len':200,'lenmode':'pixels' })
```

```python
layout = dict(title='Corona Active Case Around World',geo=dict(projection={'type':'robinson'}))
'''
type:
hammer
robinson
natural earth
mercator
stereographic
'''
```

```python
x= pg.Figure(data=[data],layout=layout)
```

```python
po.iplot(x)
```
